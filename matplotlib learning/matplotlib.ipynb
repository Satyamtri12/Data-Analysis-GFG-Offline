```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Matplotlib Basics

```python
x = np.linspace(0,5,11)
x
y = x ** 3
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
# Subplots in matplotlib

plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x ** 2, x** 2)
```

```python
plt.subplot(6,6,1)
plt.subplot(6,6,2)
plt.subplot(6,6,3)
plt.subplot(6,6,4)
```

```python
# Object Oriented Way
fig = plt.figure(figsize=(12,8), dpi =100)
axis1 = fig.add_axes([0.1,0.1,.4,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.3,.3])
axis2.plot(x,y)
```

#### Types of Plot.

#### Saving Plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("Basics_Plot.png")
```
