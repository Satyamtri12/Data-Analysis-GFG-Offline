```python
import seaborn as sns
import matplotlib.pyplot as plt
```

 #### We use distribution plot to visualize the distribution of quantitative data.
  ####  -distplot/histplot
  ####  -jointplot
  ####  -pairplot
  ####  -rugplot

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
# distplot

sns.displot(df['total_bill'], bins=20, kde=True)
```

```python
# Hist plot
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins=20, kde=True)
```

```python
# Joint Plots
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind='reg')
```

#### PairPlot

```python
sns.pairplot(df, hue='size', palette='rainbow')
```

```python
sns.pairplot(df, hue='sex', palette='rainbow')
```

#### Rugplot

```python
sns.rugplot(df)
```
