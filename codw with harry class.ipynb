```python
def greatest (a, b, c):
  if (a>b and a>c):
    return a
  elif(b>a and b>c):
    return b
  else: # This handles the case where c is the greatest or equal, or if some are equal
    return c

# The lines `a = 1`, `b = 2`, `c = 3` were misplaced inside the function's body
# after the return statements, making them unreachable. The error occurred
# because the global variables `a` (string), `b` (int), `c` (int) were passed
# to the function, leading to a type mismatch in comparison.
# We now call the function with appropriate integer values directly.
print(greatest(1, 2, 3))
```

```python

def rem(l, word):
  n = []

  for item in l:
      if not(item == word):
        n.append(item.strip(word))
  return n

l = ["Satyam", "Shubham", "Rohan", "an"]


print(rem(l, "an"))




```
