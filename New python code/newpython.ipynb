```python
# Implement a stack and queue using a list Data Type
l = []
while True:
    c = int(input('''
             1 Push Elements
             2 Pop Elements
             3 Peek Element
             4 Display stack
             5 Exit

                
              '''))
    if c==1:
        n=input("Enter the Value");
        l.append(n)
        print(l)
    elif c == 2:
        if len(l) == 0:
            print("Empty stack")
        else:    
            p=l.pop()    
            print(p)
            print(l)
    elif c==3:
        if len(l)==0:
            print("Empty Stack")
        else:
            print("Last Stack Value", l[-1])  
    elif c==4:
        print("Display Stack", 1)
    elif c==5:
        break;              


```

```python
# Initial stock price and Volatility
import random  
import datetime
initial_price = 150
volatility = 0.05


volatility_change = random.choice([-volatility, volatility])

daily_price_change = initial_price + (initial_price * volatility_change)

today = datetime.datetime.today()
print("Today's Date" , today)
print("Initial price", initial_price)
print("New price" , daily_price_change)
```

```python
# while loop
# example - Target price
current_price = 153
target_price = 155

while current_price < target_price:
    print("Price is still below the target price")
    current_price += 1

print("Now price is above the target price and you can sell the stock")
```

```python
# while else with example
price = [120,112,114,111,112]
thresold_price = 115

index = 0
while index < len(price):
    if price[index] > thresold_price:
        print(f"The price at {price[index]} breaches the thresold price {thresold_price} at {index} ")
        break
    index += 1
else:
    print("There is no price breach")    

print("search is completed")
```

```python
# For loops
# Example - Average Closing price over a 5- Days period
closing_price = [120,112,114,111,112]
total_price = 0
for i in closing_price:
    total_price += i 
avg_closing_price = total_price/ len(closing_price)
print("The average closing price for a 5 Days is", avg_closing_price)    
```
