```python
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
```

```python
arr2 = a1 + a2
arr2
```

#### Broadcasting

```python
l = [10,20,30,40]
arr = np.array(l)
```

```python
arr + 10
```

```python
arr2 = np.arange(1,26).reshape(5,5)
arr2
```

```python
arr2 + 10
```
