```python
from time import *
import random as r
def mistake(partest, usertest):
    error = 0
    for i in range(len(partest)):
        try:
            if partest[i] != usertest[i]:
                error = error + 1
        except:
            error = error +1 
    return error 
              

    
```

```python

def speed_time(time_s, time_e, userinput):
    time_delay = time_e - time_s
    time_R = round(time_delay, 2)
    speed = len(userinput)/ time_R
    return round(speed)


test = ["A paragraph is a self-contained unit of discourse in writing a particular point or idea",
" my name is Satyam Tripathi", "Welcome to wscubetech jodhpur"]

test1 = r.choice(test)
print("*****typing speed*****")
print(test1)
print()
print()
time_1 = time()
testinput = input("Enter :")
time_2 = time()

print('speed : ',speed_time(time_1, time_2, testinput),"w/sec")
print("Error : ",mistake(test1, testinput))

```
