# Series
A series is a one-dimensional labeled array capable of holding any data type. The axis labels are collectively called the index.


## Creating series

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d = {'a':10, 'b':20, 'c':30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list, index = labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
